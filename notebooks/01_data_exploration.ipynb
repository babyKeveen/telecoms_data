# 01 — Data Exploration
Initial look at the SRFG dataset schema, distributions, and trip structure.

```python
import sys
sys.path.insert(0, '/home/jovyan')
import pandas as pd
import matplotlib.pyplot as plt
from pipeline.ingest import load_raw, clean

df = clean(load_raw())
df.head()
```

```python
print(df.dtypes)
print('\nTrips:', df['trip'].nunique())
print('Date range:', df['time'].min(), 'to', df['time'].max())
print('\nNull counts:')
print(df.isnull().sum())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
df['rsrp'].hist(ax=axes[0], bins=50, color='steelblue', edgecolor='white')
axes[0].set_title('RSRP Distribution (dBm)')
axes[0].axvline(-90, color='red', linestyle='--', label='Poor threshold')
axes[0].legend()

df['sinr'].hist(ax=axes[1], bins=50, color='coral', edgecolor='white')
axes[1].set_title('SINR Distribution (dB)')
axes[1].axvline(0, color='red', linestyle='--', label='Poor threshold')
axes[1].legend()

df['signal'].value_counts().sort_index().plot(kind='bar', ax=axes[2], color='teal')
axes[2].set_title('Signal Bars Distribution')
plt.tight_layout()
```

```python
# Trips overview
trip_stats = df.groupby('trip').agg(
    records=('cell_id', 'count'),
    unique_cells=('cell_id', 'nunique'),
    avg_rsrp=('rsrp', 'mean'),
    min_rsrp=('rsrp', 'min')
).round(1)
trip_stats
```
